```python
result = []
for tag in title_tag:
  temp = {}
  # strong 태그에서 제목 추출
  strong_tag = tag.select_one('strong')
  if strong_tag:
    temp['title'] = strong_tag.get_text(strip=True)
  else:
    temp['title'] = None
  
  # onclick에서 ln_code 추출
  onclick = tag.attrs['onclick']
  ln_code = re.search(r'LN\d+', onclick)
  if ln_code:
    temp['ln_code'] = ln_code.group()
  
    
  result.append(temp)

result
```

```python
import requests
from datetime import datetime
import csv
from pathlib import Path

BASE = "https://bank.shinhan.com"
EP = f"{BASE}/serviceEndpoint/httpDigital"

HEADERS = {
    "Content-Type": "application/json;charset=UTF-8",
    "Accept": "application/json, text/javascript, */*; q=0.01",
    "Origin": BASE,
    "Referer": f"{BASE}/",
    "X-Requested-With": "XMLHttpRequest",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
}

def new_session():
    s = requests.Session()
    s.get(f"{BASE}/index.jsp", headers={"User-Agent": HEADERS["User-Agent"]}, timeout=10)
    return s

def post_json(session, payload):
    r = session.post(EP, headers=HEADERS, json=payload, timeout=20)
    r.raise_for_status()
    return r.json()

# ---------- 1) THO0409: 자동차 대출 목록 ----------
def fetch_product_page(session, page=1, pagesize=20, loan_type="40", sch_word=""):
    dataBody = {
        "ricInptRootInfo": {
            "serviceType":"TG","serviceCode":"THO0409","nextServiceCode":"",
            "skipErrorMsg":"","language":"ko",
            "callBack":"shbObj.fncDoTHO0409Callback",
            "isRule":"N","webUri":"/index.jsp"
        },
        "joinbit":"","job00":"","job10":"","job20":"","job30":"","job40":"",
        "job50":"","job60":"","job70":"","job90":"",
        "yearAmt":"","houseType":"","depositType":"","etcType":"",
        "schWord": sch_word,
        "loanType": loan_type,     # 40 = MY CAR
        "bnkType":"","salTranYn":"","shCardYn":"","topsYn":"",
        "FROM_ROW":"","END_ROW":"",
        "PAGE": page, "PAGESIZE": str(pagesize),
        "SORT_FLAG":"40","C_JUMIN_NO":""
    }
    payload = {"dataBody": dataBody,
               "dataHeader": {"trxCd":"RSRLO0201A10","language":"ko","subChannel":"49","channelGbn":"D0"}}
    j = post_json(session, payload)
    items = j.get("dataBody", {}).get("TB_PRODUCT", []) or []
    total = 0
    if items:
        try:
            total = int(items[0].get("TOTALROW","0"))
        except Exception:
            total = 0
    return items, total

def iter_products(session, loan_type="40", pagesize=20, sch_word=""):
    page = 1
    while True:
        items, total = fetch_product_page(session, page, pagesize, loan_type, sch_word)
        if not items:
            break
        for it in items:
            yield it
        if total and page * pagesize >= total:
            break
        page += 1

# ---------- 2) TDL1023: 금리 프리셋 ----------
def fetch_rate_presets(session, prod_code_numeric,
                       amount=10_000_000, period_unit="1", period="12", repay_code="2"):
    dataBody = {
        "ricInptRootInfo": {
            "serviceType":"TG","serviceCode":"TDL1023","nextServiceCode":"",
            "skipErrorMsg":"","language":"ko",
            "callBack":"shbObj.fncDoD3226",
            "isRule":"N","webUri":"/index.jsp"
        },
        # 서버가 프리셋을 만들어 돌려주니 핵심 파라미터만 넣음
        "상품CODE": str(prod_code_numeric),
        "신청금액": str(amount),
        "대출기간구분": period_unit,   # 1: 개월
        "대출기간": str(period),
        "상환방식": repay_code,        # 화면 기준 2: 원리금균등 등
    }
    payload = {"dataBody": dataBody,
               "dataHeader": {"trxCd":"RSRLO0201A07","language":"ko","subChannel":"49","channelGbn":"D0"}}
    j = post_json(session, payload)
    return j.get("dataBody", {}).get("LOAN_RATE", []) or []

# ---------- 3) D3227: 상세금리 산출 ----------
def build_d3227_payload_from_preset(preset, prod_code_numeric, amount, 기준일자):
    return {
        "dataBody": {
            "ricInptRootInfo": {
                "serviceType":"GU","serviceCode":"D3227","nextServiceCode":"",
                "skipErrorMsg":"true","language":"ko",
                "callBack":"shbObj.fncSetNewRateList",
                "exceptionCallback":"shbObj.fncDoD3227ErrCallback",
                "isRule":"N","webUri":"/index.jsp"
            },
            "거래구분":"",
            "금리기준일자": 기준일자,
            "상품CODE": str(prod_code_numeric),
            "정책금융CODE": preset.get("정책금융CODE","0"),
            "신청금액": str(amount),
            "대출기간단위CODE": preset.get("대출기간구분","1"),
            "대출기간": preset.get("대출기간","12"),
            "상환방식CODE": preset.get("상환방식","2"),
            "운용구분": preset.get("거래방식","1"),
            "등급구분": preset.get("등급구분","2"),
            "기관CODE": preset.get("기관",""),
            "조회등급": preset.get("조회등급","3"),
            "담보구분": preset.get("담보구분","7"),
            "담보대분류1": preset.get("담보종류","7"),
            "담보종류1": preset.get("담보종류1","71112"),
            "담보기준금액1": preset.get("담보금액", str(amount)),
            "지역구분코드1": preset.get("담보지역",""),
            "금리구분": preset.get("금리구분1","1"),
            "시장금리종류": preset.get("금리구분2","2"),
            "시장기간물종류": preset.get("금리구분3","6"),
            "담보기준일자": 기준일자
        },
        "dataHeader": {"trxCd":"RSRLO0201A04","language":"ko","subChannel":"49","channelGbn":"D0"}
    }

def fetch_detail_rate(session, payload):
    j = post_json(session, payload)
    db = j.get("dataBody", {})
    # R_RIBD3227_1, _2 ... 형태를 탐색
    for k, v in db.items():
        if isinstance(v, list) and k.startswith("R_RIBD3227_"):
            return v
    return []

# ---------- 엔드투엔드 ----------
def crawl_mycar_rates(loan_type="40", amount=10_000_000,
                      period_unit="1", period="12", repay_code="2", sch_word=""):
    s = new_session()
    results = []
    기준일자 = datetime.now().strftime("%Y%m%d")

    for prod in iter_products(s, loan_type=loan_type, pagesize=20, sch_word=sch_word):
        name = (prod.get("C_PROD_NAME") or "").strip()
        cprod_id = (prod.get("C_PROD_ID") or "").strip()
        prod_code_numeric = cprod_id.lstrip("S")  # D3227은 숫자코드가 안정적

        # 1) 금리 프리셋 조회
        try:
            presets = fetch_rate_presets(
                s, prod_code_numeric,
                amount=amount, period_unit=period_unit, period=period, repay_code=repay_code
            )
        except Exception as e:
            results.append({"상품명": name, "상품코드": cprod_id, "error": f"TDL1023 실패: {e}"})
            continue

        if not presets:
            results.append({"상품명": name, "상품코드": cprod_id, "note": "금리 프리셋 없음(상담전용/판매중지/비노출 가능)"})
            continue

        # 2) 프리셋들로 D3227 시도
        success = False
        last_err = None
        for p in presets:
            try:
                pay = build_d3227_payload_from_preset(p, prod_code_numeric, amount, 기준일자)
                rows = fetch_detail_rate(s, pay)
                if not rows:
                    continue
                row = rows[0]
                results.append({
                    "상품명": name,
                    "상품코드": cprod_id,
                    "프리셋_금리구분": p.get("INTEREST_NM", ""),
                    "신용등급": row.get("신용등급"),
                    "시장금리": row.get("시장금리"),
                    "우대금리": row.get("우대금리"),
                    "원가반영산출금리": row.get("원가반영산출금리"),
                    "최대결정금리": row.get("최대결정금리"),
                })
                success = True
                break
            except Exception as e:
                last_err = e
                continue

        if not success:
            results.append({"상품명": name, "상품코드": cprod_id, "error": f"D3227 산출 실패: {last_err}"})

    return results

def save_csv(rows, path="shinhan_mycar_rates.csv"):
    if not rows:
        return
    keys = sorted({k for r in rows for k in r.keys()})
    with open(path, "w", newline="", encoding="utf-8-sig") as f:
        w = csv.DictWriter(f, fieldnames=keys)
        w.writeheader()
        for r in rows:
            w.writerow(r)
    return Path(path).resolve()

if __name__ == "__main__":
    # 파라미터 조절 가능
    data = crawl_mycar_rates(
        loan_type="40",      # 자동차 대출
        amount=10_000_000,   # 신청금액
        period_unit="1",     # 개월
        period="12",         # 12개월
        repay_code="2",      # 원리금균등(화면 기준)
        sch_word=""          # 특정 키워드 필터 필요 시 입력
    )

    # 콘솔 출력
    from pprint import pprint
    pprint(data)


```

```python

import pandas as pd
from urllib.parse import urlencode

# 사전조건 점검
try:
    products
    all_results
except NameError:
    raise RuntimeError("먼저 수집(main) 셀을 실행해 products/all_results 를 생성하세요.")

# loan_product 생성
loan_products = []
seen = set()
for it in products:
    title = (it.get("C_PROD_NAME") or "").strip()
    code  = (it.get("C_PROD_ID") or "").strip()
    route = (it.get("C_INTN_URL") or "").strip()
    if route:
        url = f"{BASE}/index.jsp?cr={route}"
    elif code:
        url = f"{BASE}/index.jsp?{urlencode({'cr':'020306010000','pcd':code})}"
    else:
        url = None
    key = ("SHB", title)
    if key in seen:
        continue
    seen.add(key)
    loan_products.append({
        "product_id": len(seen),
        "bank_code": "SHB",
        "product_title": title,
        "product_code": code,
        "product_url": url,
        "is_active": 1,
    })

# loan_rates 생성 (사용자 기준 반영)
loan_rates = []
title_to_pid = { r["product_title"]: r["product_id"] for r in loan_products }

for row in all_results:
    if "최대결정금리" not in row:  # 성공 건만
        continue
    pid = title_to_pid.get(row.get("상품명", ""))
    if not pid:
        continue

    # 날짜
    ymd = (row.get("기준일자") or datetime.now().strftime("%Y%m%d")).strip()
    if len(ymd) == 8:
        effective_date = f"{ymd[:4]}-{ymd[4:6]}-{ymd[6:8]}"
    else:
        effective_date = datetime.now().strftime("%Y-%m-%d")

    # 기간(月)
    term_raw = row.get("금리기간")
    try:
        term_months = int(str(term_raw).strip()) if term_raw is not None else 0
    except Exception:
        term_months = 0

    def to_dec(x):
        if x is None: return None
        try:
            return round(float(str(x).strip()), 2)
        except Exception:
            return None

    base_rate   = to_dec(row.get("시장금리"))
    spread_rate = to_dec(row.get("최대가산금리"))
    if spread_rate is None:
        spread_rate = to_dec(row.get("가산금리"))
    pref_rate   = to_dec(row.get("우대금리"))
    min_rate    = to_dec(row.get("역마진금리"))
    max_rate    = to_dec(row.get("최대결정금리"))

    if not (pid and effective_date and term_months and base_rate is not None and spread_rate is not None and pref_rate is not None and min_rate is not None and max_rate is not None):
        continue

    source_url = next((p["product_url"] for p in loan_products if p["product_title"] == row.get("상품명")), "")

    loan_rates.append({
        "product_id": pid,
        "effective_date": effective_date,
        "term_months": term_months,
        "base_rate": f"{base_rate:.2f}",
        "spread_rate": f"{spread_rate:.2f}",
        "pref_rate": f"{pref_rate:.2f}",
        "min_rate": f"{min_rate:.2f}",
        "max_rate": f"{max_rate:.2f}",
        "source_url": source_url,
    })

# 미리보기
cols_product = ["product_id","bank_code","product_title","product_code","product_url","is_active"]
cols_rates   = ["product_id","effective_date","term_months","base_rate","spread_rate","pref_rate","min_rate","max_rate","source_url"]

df_products = pd.DataFrame(loan_products)[cols_product]
df_rates    = pd.DataFrame(loan_rates)[cols_rates]

print(f"products: {len(df_products)} rows")
display(df_products)
print(f"\nrates: {len(df_rates)} rows")
display(df_rates)

```
